| **Aspect** | **PCA** | **Factor Analysis (FA)** |
|------------|---------|---------------------------|

| **Assumes** | Components are **uncorrelated**, no model of error | Explicitly models **shared variance** + **unique/error variance** |
| **Variance** | Total variance used | Only **common/shared variance** is modeled |

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def PCA(X, isCorrMX=False):
    """
    Principal Component Analysis implementation from scratch.
    
    Parameters:
    -----------
    X : numpy.ndarray
        Data matrix where rows are observations and columns are variables
    isCorrMX : bool, default=False
        If True, use correlation matrix instead of covariance matrix
        
    Returns:
    --------
    loadings : numpy.ndarray
        Loading matrix (eigenvectors of covariance/correlation matrix)
    eigenvalues : numpy.ndarray
        Eigenvalue vector in descending order
    scores : numpy.ndarray
        Score matrix (principal components)
    """
    # Center the data
    X_centered = X - np.mean(X, axis=0)
    
    # Standardize if using correlation matrix
    if isCorrMX:
        X_std = X_centered / np.std(X_centered, axis=0, ddof=1)
    else:
        X_std = X_centered
    
    # Calculate covariance matrix or correlation matrix
    n = X.shape[0]
    cov_matrix = np.dot(X_std.T, X_std) / (n-1)
    
    # Compute eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
    
    # Sort eigenvalues and eigenvectors in descending order
    idx = eigenvalues.argsort()[::-1]
    eigenvalues = eigenvalues[idx]
    eigenvectors = eigenvectors[:, idx]
    
    # Calculate scores (principal components)
    scores = np.dot(X_std, eigenvectors)
    
    # Generate scree plot
    fig, ax1 = plt.subplots(figsize=(10, 6))
    
    # Bar plot of eigenvalues
    ax1.bar(range(1, len(eigenvalues) + 1), eigenvalues, alpha=0.7, color='b', label='Eigenvalues')
    ax1.set_xlabel('Principal Component')
    ax1.set_ylabel('Eigenvalue')
    ax1.set_xticks(range(1, len(eigenvalues) + 1))
    
    # Line plot of cumulative explained variance
    ax2 = ax1.twinx()
    total_var = np.sum(eigenvalues)
    cum_var = np.cumsum(eigenvalues) / total_var * 100
    ax2.plot(range(1, len(eigenvalues) + 1), cum_var, 'ro-', label='Cumulative Explained Variance')
    ax2.set_ylabel('Cumulative Explained Variance (%)')
    ax2.set_ylim([0, 105])
    
    # Add legends
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')
    
    plt.title('Scree Plot')
    plt.tight_layout()
    
    return eigenvectors, eigenvalues, scores

def factor_analysis(X, n_factors, max_iter=100, tol=1e-4):
    """
    Factor Analysis implementation from scratch using Maximum Likelihood Estimation.
    
    Parameters:
    -----------
    X : numpy.ndarray
        Data matrix where rows are observations and columns are variables
    n_factors : int
        Number of factors to extract
    max_iter : int, default=100
        Maximum number of iterations for the iterative algorithm
    tol : float, default=1e-4
        Tolerance for convergence
        
    Returns:
    --------
    loadings : numpy.ndarray
        Loading matrix A
    factors : numpy.ndarray
        Factor score matrix F
    communalities : numpy.ndarray
        Communality vector h_i^2
    uniquenesses : numpy.ndarray
        Uniqueness vector œà_i
    var_explained : numpy.ndarray
        Proportion of variance explained by each factor
    """
    n, p = X.shape
    
    # Standardize the data
    X_std = (X - np.mean(X, axis=0)) / np.std(X, axis=0, ddof=1)
    
    # Compute correlation matrix
    R = np.corrcoef(X_std, rowvar=False)
    
    # Initial estimate of uniquenesses (using PCA as initial guess)
    _, eigenvalues, _ = PCA(X_std, isCorrMX=True)
    
    # Initial communality estimates using Principal Component method
    h2_initial = np.zeros(p)
    for i in range(p):
        if n_factors < p:
            h2_initial[i] = np.sum(R[i, :]**2) * n_factors / p
        else:
            h2_initial[i] = 0.5  # Use 0.5 as a reasonable starting value
    
    uniquenesses = 1 - h2_initial
    uniquenesses = np.clip(uniquenesses, 0.005, 0.995)  # Ensure numerical stability
    
    # Iterative algorithm for maximum likelihood estimation
    oldunique = np.ones(p) * float('inf')
    
    for _ in range(max_iter):
        # Break if converged
        if np.sum((uniquenesses - oldunique)**2) < tol:
            break
            
        oldunique = uniquenesses.copy()
        
        # Compute adjusted correlation matrix
        diagR = np.diag(R)
        R_adj = R.copy()
        np.fill_diagonal(R_adj, diagR - uniquenesses)
        
        # Eigendecomposition of the adjusted correlation matrix
        eigenvalues, eigenvectors = np.linalg.eigh(R_adj)
        
        # Sort in descending order
        idx = eigenvalues.argsort()[::-1]
        eigenvalues = eigenvalues[idx]
        eigenvectors = eigenvectors[:, idx]
        
        # Keep only positive eigenvalues
        pos_idx = eigenvalues > 0
        pos_eigenvalues = eigenvalues[pos_idx][:n_factors]
        pos_eigenvectors = eigenvectors[:, pos_idx][:, :n_factors]
        
        # Calculate loadings matrix
        loadings = pos_eigenvectors @ np.diag(np.sqrt(pos_eigenvalues))
        
        # Update uniquenesses
        communalities = np.sum(loadings**2, axis=1)
        uniquenesses = 1 - communalities
        uniquenesses = np.clip(uniquenesses, 0.005, 0.995)  # Ensure numerical stability
    
    # Calculate factor scores using regression method
    Phi = np.eye(n_factors)  # Factor correlation matrix (identity for orthogonal factors)
    diagU = np.diag(uniquenesses)
    inv_diag_sqrt_U = np.diag(1 / np.sqrt(uniquenesses))
    
    # Factor score coefficients
    B = np.linalg.inv(loadings.T @ np.linalg.inv(diagU) @ loadings + Phi) @ loadings.T @ np.linalg.inv(diagU)
    
    # Factor scores
    factors = X_std @ B.T
    
    # Calculate communalities
    communalities = np.sum(loadings**2, axis=1)
    
    # Calculate variance explained by each factor
    var_explained = np.sum(loadings**2, axis=0) / p
    
    return loadings, factors, communalities, uniquenesses, var_explained

def load_auto_mpg():
    """
    Load Auto MPG dataset from file
    """
    # Define column names
    cols = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 
            'acceleration', 'model_year', 'origin', 'car_name']
    
    # Load the data
    df = pd.read_csv('/Users/chun-feitan/Desktop/tpe/IE5054/HW/auto-mpg.data.txt', delim_whitespace=True, names=cols)
    
    # Convert horsepower to numeric, handling '?' as NaN
    df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
    
    # Drop NaN values
    df = df.dropna()
    
    # Drop car_name as it's a unique string for each instance
    df = df.drop(columns=['car_name'])
    
    return df

def apply_fa_to_auto_mpg():
    """
    Apply Factor Analysis to Auto MPG dataset
    """
    # Load and preprocess the data
    df = load_auto_mpg()
    
    # Print dataset information
    print("Auto MPG Dataset Information:")
    print(f"Number of instances: {df.shape[0]}")
    print(f"Number of attributes: {df.shape[1]}")
    print("\nFirst 5 rows:")
    print(df.head())
    
    # Select numerical features for FA
    X = df.values
    
    # Apply factor analysis with 2 factors
    loadings, factors, communalities, uniquenesses, var_explained = factor_analysis(X, n_factors=2)
    
    # Display results
    print("\nFactor Analysis Results:")
    print("\nLoading Matrix (A):")
    cols = df.columns
    loadings_df = pd.DataFrame(loadings, index=cols, columns=[f'Factor {i+1}' for i in range(loadings.shape[1])])
    print(loadings_df)
    
    print("\nCommunalities (h¬≤):")
    comm_df = pd.DataFrame({'Communality': communalities}, index=cols)
    print(comm_df)
    
    print("\nUniquenesses (œà):")
    uniq_df = pd.DataFrame({'Uniqueness': uniquenesses}, index=cols)
    print(uniq_df)
    
    print("\nProportion of Variance Explained:")
    var_df = pd.DataFrame({'Proportion': var_explained}, index=[f'Factor {i+1}' for i in range(len(var_explained))])
    print(var_df)
    print(f"Total Variance Explained: {np.sum(var_explained):.4f}")
    
    # Create a factor loadings plot
    plt.figure(figsize=(10, 6))
    plt.scatter(loadings[:, 0], loadings[:, 1], s=100)
    
    # Add variable names as labels
    for i, var in enumerate(cols):
        plt.annotate(var, (loadings[i, 0], loadings[i, 1]), 
                    xytext=(5, 5), textcoords='offset points')
    
    # Add coordinate lines
    plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
    
    plt.title('Factor Loadings Plot')
    plt.xlabel('Factor 1')
    plt.ylabel('Factor 2')
    plt.grid(alpha=0.3)
    plt.tight_layout()
    
    # Return the first few rows of the factor scores
    factor_scores_df = pd.DataFrame(factors, columns=[f'Factor {i+1}' for i in range(factors.shape[1])])
    print("\nFirst 5 rows of Factor Scores:")
    print(factor_scores_df.head())
    
    plt.show()
    
    return loadings, factors, communalities, uniquenesses, var_explained

# Run the analysis when the script is executed
if __name__ == "__main__":
    # Create a sample dataset for testing PCA
    np.random.seed(42)
    X_test = np.random.randn(100, 5)
    X_test[:, 1] = X_test[:, 0] * 0.9 + np.random.randn(100) * 0.1  # Create correlation
    X_test[:, 2] = X_test[:, 0] * 0.8 + np.random.randn(100) * 0.2
    
    print("Testing PCA function:")
    loadings, eigenvalues, scores = PCA(X_test, isCorrMX=True)
    print(f"Eigenvalues: {eigenvalues}")
    print(f"Loadings shape: {loadings.shape}")
    print(f"Scores shape: {scores.shape}")
    
    print("\nApplying Factor Analysis to Auto MPG dataset:")
    apply_fa_to_auto_mpg()
```

```python
import pandas as pd
import numpy as np
from sklearn.cross_decomposition import PLSRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from math import sqrt
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Column names
cols = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight',
        'acceleration', 'model_year', 'origin', 'car_name']

# Load the dataset
df = pd.read_csv(
    '/Users/chun-feitan/Desktop/tpe/IE5054/HW/auto-mpg.data.txt',
    sep='\s+', names=cols
)

# Handle missing values in horsepower
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
df['horsepower'].fillna(df['horsepower'].mean(), inplace=True)

# Feature matrix and response variables
X = df.drop(['mpg', 'model_year', 'car_name'], axis=1)
y_univariate = df[['mpg']]
y_multivariate = df[['mpg', 'model_year']]

# Split data into training and testing sets (300 train / 92 test)
X_train, X_test, y_uni_train, y_uni_test = train_test_split(
    X, y_univariate, train_size=300, random_state=42
)
_, _, y_multi_train, y_multi_test = train_test_split(
    X, y_multivariate, train_size=300, random_state=42
)

# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# --- Function to train and evaluate PLSR ---
def train_evaluate_plsr(X_train, X_test, y_train, y_test, max_components=10):
    # Automatically cap based on allowed dimensions
    max_possible = min(X_train.shape[1], y_train.shape[1] if len(y_train.shape) > 1 else 1)
    max_components = min(max_components, max_possible)

    mse_list = []
    for n in range(1, max_components + 1):
        pls = PLSRegression(n_components=n)
        pls.fit(X_train, y_train)
        y_pred = pls.predict(X_test)
        mse_list.append(mean_squared_error(y_test, y_pred))
    best_n = np.argmin(mse_list) + 1
    pls_final = PLSRegression(n_components=best_n)
    pls_final.fit(X_train, y_train)
    y_pred_final = pls_final.predict(X_test)
    return pls_final, y_pred_final, best_n, mse_list

# --- Part A: Univariate PLSR ---
pls_uni, y_uni_pred, optimal_components_uni, mse_values_uni = train_evaluate_plsr(
    X_train_scaled, X_test_scaled, y_uni_train, y_uni_test
)

mse_uni = mean_squared_error(y_uni_test, y_uni_pred)
rmse_uni = sqrt(mse_uni)
r2_uni = r2_score(y_uni_test, y_uni_pred)

print(f"Part A: Univariate PLS Regression (y = mpg)")
print(f"Optimal Components: {optimal_components_uni}")
print(f"MSE: {mse_uni:.4f}, RMSE: {rmse_uni:.4f}, R¬≤: {r2_uni:.4f}")

# --- Part B: Multivariate PLSR ---
pls_multi, y_multi_pred, optimal_components_multi, mse_values_multi = train_evaluate_plsr(
    X_train_scaled, X_test_scaled, y_multi_train, y_multi_test
)

# Split predictions
mpg_actual = y_multi_test['mpg'].values
model_year_actual = y_multi_test['model_year'].values
mpg_pred_multi = y_multi_pred[:, 0]
model_year_pred = y_multi_pred[:, 1]

# MPG evaluation
mse_mpg_multi = mean_squared_error(mpg_actual, mpg_pred_multi)
rmse_mpg_multi = sqrt(mse_mpg_multi)
r2_mpg_multi = r2_score(mpg_actual, mpg_pred_multi)

# Model Year evaluation
mse_model_year = mean_squared_error(model_year_actual, model_year_pred)
rmse_model_year = sqrt(mse_model_year)
r2_model_year = r2_score(model_year_actual, model_year_pred)

print(f"\nPart B: Multivariate PLS Regression (y = [mpg, model_year])")
print(f"Optimal Components: {optimal_components_multi}")
print(f"MPG ‚Üí MSE: {mse_mpg_multi:.4f}, RMSE: {rmse_mpg_multi:.4f}, R¬≤: {r2_mpg_multi:.4f}")
print(f"Model Year ‚Üí MSE: {mse_model_year:.4f}, RMSE: {rmse_model_year:.4f}, R¬≤: {r2_model_year:.4f}")

# --- Comparison ---
print("\nüìä MPG Prediction Comparison:")
print(f"Univariate ‚Üí RMSE: {rmse_uni:.4f}, R¬≤: {r2_uni:.4f}")
print(f"Multivariate ‚Üí RMSE: {rmse_mpg_multi:.4f}, R¬≤: {r2_mpg_multi:.4f}")
print(f"RMSE Difference: {rmse_uni - rmse_mpg_multi:.4f}")
print(f"Percentage Improvement: {(1 - rmse_mpg_multi/rmse_uni) * 100:.2f}%")

# --- Plots ---
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(mse_values_uni) + 1), mse_values_uni, marker='o', label='Univariate')
plt.plot(range(1, len(mse_values_multi) + 1), mse_values_multi, marker='s', label='Multivariate')
plt.title('MSE vs. Number of Components')
plt.xlabel('Number of Components')
plt.ylabel('Mean Squared Error')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Actual vs Predicted MPG
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(y_uni_test, y_uni_pred, alpha=0.7)
plt.plot([y_uni_test.min(), y_uni_test.max()], [y_uni_test.min(), y_uni_test.max()], 'r--')
plt.title("Univariate: Actual vs Predicted MPG")
plt.xlabel("Actual MPG")
plt.ylabel("Predicted MPG")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.scatter(mpg_actual, mpg_pred_multi, alpha=0.7)
plt.plot([mpg_actual.min(), mpg_actual.max()], [mpg_actual.min(), mpg_actual.max()], 'r--')
plt.title("Multivariate: Actual vs Predicted MPG")
plt.xlabel("Actual MPG")
plt.ylabel("Predicted MPG")
plt.grid(True)

plt.tight_layout()
plt.show()

# Residual plots
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
residuals_uni = y_uni_test.values.flatten() - y_uni_pred.flatten()
plt.scatter(y_uni_pred, residuals_uni, alpha=0.7)
plt.axhline(y=0, color='red', linestyle='--')
plt.title("Univariate: Residual Plot")
plt.xlabel("Predicted MPG")
plt.ylabel("Residuals")
plt.grid(True)

plt.subplot(1, 2, 2)
residuals_multi = mpg_actual - mpg_pred_multi
plt.scatter(mpg_pred_multi, residuals_multi, alpha=0.7)
plt.axhline(y=0, color='red', linestyle='--')
plt.title("Multivariate: Residual Plot (MPG)")
plt.xlabel("Predicted MPG")
plt.ylabel("Residuals")
plt.grid(True)

plt.tight_layout()
plt.show()

# Model year prediction plot
plt.figure(figsize=(6, 6))
plt.scatter(model_year_actual, model_year_pred, alpha=0.7)
plt.plot([min(model_year_actual), max(model_year_actual)],
         [min(model_year_actual), max(model_year_actual)], 'r--')
plt.title("Multivariate: Actual vs Predicted Model Year")
plt.xlabel("Actual Model Year")
plt.ylabel("Predicted Model Year")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from sklearn.decomposition import FactorAnalysis
from scipy.stats import pearsonr

# === Parameters ===
DATA_DIR = "/Users/chun-feitan/Desktop/tpe/IE5054/ORL Faces"
IMG_SIZE = (56, 46)  # height, width

# Full Gender labels for all 40 subjects
full_gender_labels = {
    "01": 0, "02": 1, "03": 1, "04": 1, "05": 1,
    "06": 1, "07": 1, "08": 0, "09": 1, "10": 0,
    "11": 1, "12": 1, "13": 1, "14": 1, "15": 1,
    "16": 1, "17": 1, "18": 1, "19": 1, "20": 1,
    "21": 1, "22": 1, "23": 1, "24": 1, "25": 1,
    "26": 1, "27": 1, "28": 1, "29": 1, "30": 1,
    "31": 1, "32": 0, "33": 1, "34": 1, "35": 1,
    "36": 1, "37": 1, "38": 1, "39": 1, "40": 1,
}
# 0: female (f), 1: male (m)

# === Load Data ===
X = []
y = []

for filename in sorted(os.listdir(DATA_DIR)):
    if not filename.endswith(".png"):
        continue

    person_id = filename[:2]
    if person_id not in full_gender_labels:
        print(f"Warning: Gender label not found for person ID: {person_id}")
        continue

    img_path = os.path.join(DATA_DIR, filename)
    img = Image.open(img_path).convert("L")  # grayscale
    img_array = np.array(img).flatten()     # 1D vector

    X.append(img_array)
    y.append(full_gender_labels[person_id])

X = np.array(X)
y = np.array(y)

print(f"Loaded {X.shape} images, each of dimension {X.shape[1]} pixels.")

# Transpose X to get a 2576 √ó 400 data matrix as specified
X_transposed = X.T  # Shape: (2576, 400)
```

```python
from sklearn.decomposition import PCA

# Apply PCA to get explained variance ratio as a proxy
pca = PCA()
pca.fit(X_transposed.T)  # shape (400, 2576)

# Cumulative variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

# Check how many components are needed for each threshold
thresholds = [0.5, 0.6, 0.7, 0.8, 0.9]
components_required = {}

for threshold in thresholds:
    num_components = np.argmax(cumulative_variance >= threshold) + 1
    components_required[f"{int(threshold * 100)}%"] = num_components

print("Number of components required:")
for k, v in components_required.items():
    print(f"{k}: {v} components")
```

```python
from sklearn.decomposition import FactorAnalysis
from sklearn.preprocessing import MinMaxScaler

# Use number of components that explains at least 80% variance
n_factors = components_required["80%"]

fa = FactorAnalysis(n_components=n_factors)
fa.fit(X_transposed.T)

# Extract the first factor (loading vector)
first_factor = fa.components_[0]  # Shape: (2576,)

# Rescale to [0, 255]
scaler = MinMaxScaler(feature_range=(0, 255))
first_factor_scaled = scaler.fit_transform(first_factor.reshape(-1, 1)).reshape(2576,)

# Reshape into 46 x 56 (width x height, because of how it was flattened)
factor_img = first_factor_scaled.reshape(56, 46).T  # transpose to get original orientation

# Visualize
plt.imshow(factor_img, cmap='gray')
plt.title("First Extracted Factor (Rescaled)")
plt.axis('off')
plt.show()
```
