## 三、 AI 如何應用在我的專業領域

### 1. 預計完成的專案/研究

我計畫開發一個 **「基於 Agentic Network 的跨界策展系統」**。

* **核心目標**：利用多個 AI Agent 分別模擬「文學評論家」、「藝術史學家」與「創意總監」的角色，針對特定的人文主題進行風格混搭與數位創作。
* **技術結合**：結合課程中的 **Context Engineering (RAG)** 與 **多模態 (Multimodal)** 技術，減少 AI 幻覺並提升輸出的美學價值。

### 2. 這門課對我的幫助

課程大綱中的以下主題對我的職涯與學術目標非常有幫助：

* **推理模型與評測 (第 7 週)**：幫助我了解如何量化評估人文生成內容的品質。
* **數位分身的社會學 (第 14 週)**：讓我思考在開發 AI 代理時，如何處理人類價值對齊與倫理邊界。

## 四、 本份作業使用的 Markdown 語法

1. **標題 (Headings)**：使用不同層級標題區分章節。
2. **表格 (Tables)**：完整呈現多校修課清單。
3. **粗體 (Bold)**：強調關鍵課程與個人亮點。
4. **引用 (Blockquotes)**：標註技術參考來源。
5. **連結 (Links)**：提供相關資源的快速連結。
6. **無序清單 (Lists)**：結構化陳述背景資訊。
