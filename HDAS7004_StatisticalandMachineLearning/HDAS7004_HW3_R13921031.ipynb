## ü§î Problem 1.
Ë´ãÂ∞éÂá∫ridge regressionÁöÑËø¥Ê≠∏‰øÇÊï∏Ëß£  
$\hat{\beta}^{\text{ridge}} = (X^T * X + \lambda * I)^{-1} * X^{Y} * Y$

To derive the ridge regression coefficient solution step-by-step, we start with the ridge regression objective function on page 12 of the lecture note 3.2.  
$J(\beta) = RSS + \lambda \sum_{j=1,2,...p} \beta_{j}^2 = \sum_{i=1}^{n} (y_i - \beta_0 - \sum_{j=1}^{p} \beta_j x_{ij})^2 + \lambda \sum_{j=1}^{p} \beta_j^2$

### ‚úÖ STEP1. Express in Matrix Form:
We want to rewrite the above function in matrix form. First, let‚Äôs define the following matrices and vectors:

- Let $ \mathbf{X} $ be the $ n \times p $ matrix of predictors where each row corresponds to a data point and each column corresponds to a predictor variable, i.e., $ \mathbf{X} = \begin{pmatrix} \mathbf{x}_1^\top \\ \mathbf{x}_2^\top \\ \vdots \\ \mathbf{x}_n^\top \end{pmatrix} $.
- Let $ \mathbf{y} $ be the $ n \times 1 $ vector of observed responses, i.e., $ \mathbf{y} = \begin{pmatrix} y_1 \\ y_2 \\ \vdots \\ y_n \end{pmatrix} $.
- Let $ \boldsymbol{\beta} $ be the $ p \times 1 $ vector of regression coefficients excluding the intercept term (we will assume the intercept $ \beta_0 $ is handled separately or included in $ \mathbf{X} $ by adding a column of ones).
- $ \lambda $ is the regularization parameter.
- $ I $ is the $ p \times p $ identity matrix.

We can rewrite the objective function as:

$
J(\boldsymbol{\beta}) = \| \mathbf{y} - \mathbf{X} \boldsymbol{\beta} \|_2^2 + \lambda \|\boldsymbol{\beta}\|_2^2
$

Where:

- $ \| \mathbf{y} - \mathbf{X} \boldsymbol{\beta} \|_2^2 = (\mathbf{y} - \mathbf{X} \boldsymbol{\beta})^\top (\mathbf{y} - \mathbf{X} \boldsymbol{\beta}) $ is the residual sum of squares (RSS).
- $ \|\boldsymbol{\beta}\|_2^2 = \boldsymbol{\beta}^\top \boldsymbol{\beta} $ is the L2 regularization term.

Now, expand the first term:

$
\| \mathbf{y} - \mathbf{X} \boldsymbol{\beta} \|_2^2 = (\mathbf{y} - \mathbf{X} \boldsymbol{\beta})^\top (\mathbf{y} - \mathbf{X} \boldsymbol{\beta}) = \mathbf{y}^\top \mathbf{y} - 2 \mathbf{y}^\top \mathbf{X} \boldsymbol{\beta} + \boldsymbol{\beta}^\top \mathbf{X}^\top \mathbf{X} \boldsymbol{\beta}
$

Thus, the objective function becomes:

$
J(\boldsymbol{\beta}) = \mathbf{y}^\top \mathbf{y} - 2 \mathbf{y}^\top \mathbf{X} \boldsymbol{\beta} + \boldsymbol{\beta}^\top \mathbf{X}^\top \mathbf{X} \boldsymbol{\beta} + \lambda \boldsymbol{\beta}^\top \boldsymbol{\beta}
$

### ‚úÖ STEP2. Take the Gradient with Respect to $ \boldsymbol{\beta} $:
Next, we take the gradient of the objective function with respect to $ \boldsymbol{\beta} $. We will differentiate each term one by one.

- The gradient of $ \mathbf{y}^\top \mathbf{y} $ with respect to $ \boldsymbol{\beta} $ is zero because it does not depend on $ \boldsymbol{\beta} $.
- The gradient of $ -2 \mathbf{y}^\top \mathbf{X} \boldsymbol{\beta} $ is $ -2 \mathbf{X}^\top \mathbf{y} $.
- The gradient of $ \boldsymbol{\beta}^\top \mathbf{X}^\top \mathbf{X} \boldsymbol{\beta} $ is $ 2 \mathbf{X}^\top \mathbf{X} \boldsymbol{\beta} $.
- The gradient of $ \lambda \boldsymbol{\beta}^\top \boldsymbol{\beta} $ is $ 2 \lambda \boldsymbol{\beta} $.

## üìù Proof of step 2.

### 2-1. Gradient of $ -2 \mathbf{y}^\top \mathbf{X} \boldsymbol{\beta} $
The term $ -2 \mathbf{y}^\top \mathbf{X} \boldsymbol{\beta} $ represents a linear form in $ \boldsymbol{\beta} $.   
To compute the gradient of this term, recall that the gradient of a linear function $ \mathbf{a}^\top \mathbf{x} $ with respect to $ \mathbf{x} $ is simply the vector $ \mathbf{a} $.

- The term $ -2 \mathbf{y}^\top \mathbf{X} \boldsymbol{\beta} $ is linear in $ \boldsymbol{\beta} $, where $ \mathbf{y}^\top \mathbf{X} $ is a constant vector with respect to $ \boldsymbol{\beta} $.
- So, the gradient of $ -2 \mathbf{y}^\top \mathbf{X} \boldsymbol{\beta} $ with respect to $ \boldsymbol{\beta} $ is $ -2 \mathbf{X}^\top \mathbf{y} $, because $\nabla_{\boldsymbol{\beta}} (-2 \mathbf{y}^\top \mathbf{X} \boldsymbol{\beta}) = -2 \mathbf{X}^\top \mathbf{y}$

### 2-2. Gradient of $ \boldsymbol{\beta}^\top \mathbf{X}^\top \mathbf{X} \boldsymbol{\beta} $
This term represents a quadratic form in $ \boldsymbol{\beta} $. We need to apply the known result for the gradient of a quadratic form $ \boldsymbol{\beta}^\top A \boldsymbol{\beta} $, where $ A $ is a symmetric matrix.

- In this case, $ A = \mathbf{X}^\top \mathbf{X} $, which is symmetric.
- The gradient of $ \boldsymbol{\beta}^\top A \boldsymbol{\beta} $ with respect to $ \boldsymbol{\beta} $ is $ 2 A \boldsymbol{\beta} $.
- Therefore, the gradient of $ \boldsymbol{\beta}^\top \mathbf{X}^\top \mathbf{X} \boldsymbol{\beta} $ with respect to $ \boldsymbol{\beta} $ is $\nabla_{\boldsymbol{\beta}} (\boldsymbol{\beta}^\top \mathbf{X}^\top \mathbf{X} \boldsymbol{\beta}) = 2 \mathbf{X}^\top \mathbf{X} \boldsymbol{\beta}$

### 2-3. Gradient of $ \lambda \boldsymbol{\beta}^\top \boldsymbol{\beta} $
This term also represents a quadratic form in $ \boldsymbol{\beta} $, specifically $ \lambda \boldsymbol{\beta}^\top \boldsymbol{\beta} $, where $ \lambda $ is a scalar constant.

- The gradient of $ \boldsymbol{\beta}^\top \boldsymbol{\beta} $ with respect to $ \boldsymbol{\beta} $ is $ 2 \boldsymbol{\beta} $, because $\nabla_{\boldsymbol{\beta}} (\boldsymbol{\beta}^\top \boldsymbol{\beta}) = 2 \boldsymbol{\beta}$

- So, the gradient of $ \lambda \boldsymbol{\beta}^\top \boldsymbol{\beta} $ with respect to $ \boldsymbol{\beta} $ is $\nabla_{\boldsymbol{\beta}} (\lambda \boldsymbol{\beta}^\top \boldsymbol{\beta}) = 2 \lambda \boldsymbol{\beta}$

### Summary of the proof.

- The gradient of $ -2 \mathbf{y}^\top \mathbf{X} \boldsymbol{\beta} $ with respect to $ \boldsymbol{\beta} $ is $ -2 \mathbf{X}^\top \mathbf{y} $ because it's linear in $ \boldsymbol{\beta} $.
- The gradient of $ \boldsymbol{\beta}^\top \mathbf{X}^\top \mathbf{X} \boldsymbol{\beta} $ with respect to $ \boldsymbol{\beta} $ is $ 2 \mathbf{X}^\top \mathbf{X} \boldsymbol{\beta} $ because it is a quadratic form.
- The gradient of $ \lambda \boldsymbol{\beta}^\top \boldsymbol{\beta} $ with respect to $ \boldsymbol{\beta} $ is $ 2 \lambda \boldsymbol{\beta} $ because it's a scaled quadratic form.

Integrate the above information, we can conclude that the gradient of the objective function is:

$
\nabla_{\boldsymbol{\beta}} J(\boldsymbol{\beta}) = -2 \mathbf{X}^\top \mathbf{y} + 2 \mathbf{X}^\top \mathbf{X} \boldsymbol{\beta} + 2 \lambda \boldsymbol{\beta}
$

or

$
\nabla_{\boldsymbol{\beta}} J(\boldsymbol{\beta}) = 2 \mathbf{X}^\top \mathbf{X} \boldsymbol{\beta} - 2 \mathbf{X}^\top \mathbf{y} + 2 \lambda \boldsymbol{\beta}
$

### ‚úÖ STEP3. Set the Gradient Equal to Zero:
To find the minimizer, set the gradient equal to zero:

$
2 \mathbf{X}^\top \mathbf{X} \boldsymbol{\beta} - 2 \mathbf{X}^\top \mathbf{y} + 2 \lambda \boldsymbol{\beta} = 0
$

Simplify:

$
\mathbf{X}^\top \mathbf{X} \boldsymbol{\beta} + \lambda \boldsymbol{\beta} = \mathbf{X}^\top \mathbf{y}
$

Factor out $ \boldsymbol{\beta} $:

$
(\mathbf{X}^\top \mathbf{X} + \lambda I) \boldsymbol{\beta} = \mathbf{X}^\top \mathbf{y}
$

### ‚úÖ STEP4. Solve for $ \boldsymbol{\beta} $:
Finally, solve for $ \boldsymbol{\beta} $:

$
\boldsymbol{\beta}^{\text{ridge}} = (\mathbf{X}^\top \mathbf{X} + \lambda I)^{-1} \mathbf{X}^\top \mathbf{y}
$

This is the solution for the ridge regression coefficients.

## ü§î Problem 2.
Ë´ãË≠âÊòé $MSE(\hat{\beta})=Var(\hat{\beta})+Bias^2(\hat{\beta})$

### ‚ñ∂Ô∏è ÂèÉËÄÉ YouTube Ë¶ñÈ†ª 
[[Proof] MSE = Variance + Bias¬≤](https://youtu.be/qmVT35gFW-Q?si=PSZRS_P1lWRfYstK)

To prove that $MSE(\hat{\beta}) = Var(\hat{\beta}) + Bias^2(\hat{\beta})$, we can use the definition of Mean Squared Error (MSE), variance, and bias.

### ‚úÖ Step 1: Define MSE, Variance, and Bias
- **MSE**: The Mean Squared Error of an estimator $\hat{\beta}$ is defined as $MSE(\hat{\beta}) = E[(\hat{\beta} - \beta)^2]$, where $\beta$ is the true parameter.
- **Variance**: The variance of $\hat{\beta}$ is $Var(\hat{\beta}) = E[(\hat{\beta} - E[\hat{\beta}])^2]$.
- **Bias**: The bias of $\hat{\beta}$ is $Bias(\hat{\beta}) = E[\hat{\beta}] - \beta$.

### ‚úÖ Step 2: Derive the Relationship
Starting with the definition of MSE: $MSE(\hat{\beta}) = E[(\hat{\beta} - \beta)^2]$

We can expand this expression by adding and subtracting $E[\hat{\beta}]$: $MSE(\hat{\beta}) = E[(\hat{\beta} - E[\hat{\beta}] + E[\hat{\beta}] - \beta)^2]$

Expanding the square gives:
$MSE(\hat{\beta}) = E[(\hat{\beta} - E[\hat{\beta}])^2 + 2(\hat{\beta} - E[\hat{\beta}])(E[\hat{\beta}] - \beta) + (E[\hat{\beta}] - \beta)^2]$

### ‚úÖ Step 3: Simplify Using Expectation Properties
Since $E[\hat{\beta}]$ is a constant, and using the linearity of expectation:
- $E[(\hat{\beta} - E[\hat{\beta}])^2] = Var(\hat{\beta})$
- $E[2(\hat{\beta} - E[\hat{\beta}])(E[\hat{\beta}] - \beta)] = 2(E[\hat{\beta}] - E[\hat{\beta}])(E[\hat{\beta}] - \beta) = 0$, because $E[\hat{\beta} - E[\hat{\beta}]] = 0$.
    - üìù Please see "Proof of the second term" for further explanation.
- $E[(E[\hat{\beta}] - \beta)^2] = (E[\hat{\beta}] - \beta)^2 = Bias^2(\hat{\beta})$
    - $E[\hat{\beta}]$ is the expected value of the estimator $\hat{\beta}$. It is a constant, not a random variable.
    - The term $(E[\hat{\beta}] - \beta)^2$ is simply a constant, since both $E[\hat{\beta}]$ and $\beta$ are constants.
    - When you compute $E[(E[\hat{\beta}] - \beta)^2]$, it is the expected value of a constant, which is just the constant itself because the expectation of a constant is the constant.

Thus, the equation simplifies to:
$MSE(\hat{\beta}) = Var(\hat{\beta}) + Bias^2(\hat{\beta})$

This proves the relationship between MSE, variance, and bias of an estimator.

### üìù Proof of the second term.
Left-Hand-Side: $ E[2(\hat{\beta} - E[\hat{\beta}])(E[\hat{\beta}] - \beta)] $

- The term $ \hat{\beta} - E[\hat{\beta}] $ represents the **deviation** of the estimator $\hat{\beta}$ from its expected value $E[\hat{\beta}]$. This deviation is a random variable with mean zero.
- The term $ E[\hat{\beta}] - \beta $ is a constant, as $E[\hat{\beta}]$ and $\beta$ are both constants.

So the expression inside the expectation is the product of two terms:
1. $ (\hat{\beta} - E[\hat{\beta}]) $, which is a random variable with mean zero.
2. $ (E[\hat{\beta}] - \beta) $, which is a constant.

When you take the expectation of this product, you're essentially calculating the expectation of the product of a random variable with mean zero and a constant.

$
E[2(\hat{\beta} - E[\hat{\beta}]) \cdot (E[\hat{\beta}] - \beta)] = E[(\hat{\beta} - E[\hat{\beta}])]\cdot 2(E[\hat{\beta}] - \beta)
= (0)\cdot 2(E[\hat{\beta}] - \beta) = 0
$

### ‚úÖ Conclusion
 - The Mean Squared Error (MSE) of an estimator $\hat{\beta}$ can be decomposed into the variance of the estimator plus the squared bias of the estimator.
 - This decomposition is fundamental in understanding the trade-offs in model selection and regularization in linear models.

### Handling Different Feature Types in Shrinkage Methods

#### 1. Numerical Features

For numerical features like 'Age', 'Sleep Duration', and 'Heart Rate':

- **Scaling is essential**: Shrinkage methods are sensitive to the scale of features. Without scaling, features with larger values will be penalized more heavily.
  
- **Standard scaling**: Convert to zero mean and unit variance using `StandardScaler()` as shown in the code. This ensures all numerical features contribute equally to the regularization penalty.

- **Why it matters**: If we didn't scale, variables measured in larger units (like 'Daily Steps' which is in thousands) would be penalized more severely than variables measured in smaller units (like 'Sleep Duration' in hours) regardless of their actual importance.

#### 2. Categorical Features

For categorical features like 'Gender', 'Occupation', and 'Sleep Disorder':

- **One-hot encoding or dummy variables**: Convert categorical variables into binary indicators. The code uses `pd.get_dummies()` to create these binary features.

- **Handling multicollinearity**: When creating dummy variables, we typically use `drop_first=True` to avoid perfect multicollinearity (the "dummy variable trap").

- **Group-wise selection**: One limitation of standard shrinkage methods is they treat each dummy variable independently, potentially selecting some levels of a categorical variable while excluding others.

Example from our data:
```python
# Create dummy variables
df_encoded = pd.get_dummies(df, columns=['Gender', 'Occupation', 'BMI Category', 'Sleep Disorder'], drop_first=True)
```

#### Special Considerations for Shrinkage Methods

1. **Ridge Regression**: 
   - Does not perform feature selection
   - Shrinks coefficients toward zero but rarely to exactly zero
   - All dummy variables from categorical features remain in the model
   - Handles multicollinearity well

2. **LASSO**:
   - Performs feature selection by setting some coefficients to exactly zero
   - May arbitrarily select one level of a categorical variable while excluding others
   - When two features are highly correlated, LASSO tends to select one and ignore the other

3. **Elastic Net**:
   - Combines Ridge and LASSO penalties
   - Better than LASSO at handling groups of correlated features
   - Can select entire groups of dummy variables more consistently than LASSO

#### Advanced Techniques for Categorical Variables (I did not implement these advanced techniques.)

For more principled handling of categorical variables in shrinkage methods, consider:

1. **Group LASSO**: Applies the penalty to groups of variables (all dummies from one categorical variable) rather than individual coefficients.

2. **Sparse Group LASSO**: Combines overall sparsity with group-level selection.

3. **Categorical Encoding Alternatives**: Methods like target encoding can sometimes be more effective than one-hot encoding for high-cardinality categorical variables.

````python
# Subset Selection Methods for Sleep Quality Prediction

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import RFECV
from itertools import combinations
import time

# Assuming df_encoded, X, y are already defined from previous code
# And X_train, X_test, y_train, y_test are already split
# And X_train_scaled, X_test_scaled are already scaled

# For demonstration, we'll use a smaller subset of features if there are many
# This is because Best Subset becomes computationally expensive with many features
# Let's select top features based on correlation with the target

def calculate_feature_importance(X, y):
    """Calculate feature importance based on correlation with target"""
    importance = {}
    for col in X.columns:
        importance[col] = abs(np.corrcoef(X[col], y)[0, 1])
    return importance

# Calculate feature importance
importance = calculate_feature_importance(X_train, y_train)
importance_df = pd.DataFrame({'Feature': list(importance.keys()), 
                             'Importance': list(importance.values())})
importance_df = importance_df.sort_values('Importance', ascending=False)

# Select top 15 features for subset selection (adjust as needed)
top_features = importance_df['Feature'].head(15).tolist()
X_train_top = X_train[top_features]
X_test_top = X_test[top_features]

# Scale these features
scaler_top = StandardScaler()
X_train_top_scaled = scaler_top.fit_transform(X_train_top)
X_test_top_scaled = scaler_top.transform(X_test_top)

# Function to evaluate a model with given features
def evaluate_model(X_train, X_test, y_train, y_test, feature_indices):
    # Extract selected features
    X_train_selected = X_train[:, feature_indices]
    X_test_selected = X_test[:, feature_indices]
    
    # Fit model
    model = LinearRegression()
    model.fit(X_train_selected, y_train)
    
    # Evaluate
    train_pred = model.predict(X_train_selected)
    test_pred = model.predict(X_test_selected)
    
    train_mse = mean_squared_error(y_train, train_pred)
    test_mse = mean_squared_error(y_test, test_pred)
    train_r2 = r2_score(y_train, train_pred)
    test_r2 = r2_score(y_test, test_pred)
    
    return {
        'train_mse': train_mse,
        'test_mse': test_mse,
        'train_r2': train_r2,
        'test_r2': test_r2,
        'feature_indices': feature_indices,
        'n_features': len(feature_indices)
    }

# 4. Best Subset Selection
def best_subset_selection(X_train, X_test, y_train, y_test, max_features=15):
    n_features = X_train.shape[1]
    max_features = min(max_features, n_features)
    
    results = []
    best_model_per_size = []
    
    print(f"Running Best Subset Selection (may take time for {n_features} features)...")
    start_time = time.time()
    
    for k in range(1, max_features + 1):
        best_score = -np.inf
        best_feature_set = None
        
        # Generate all combinations of k features
        for feature_indices in combinations(range(n_features), k):
            # Evaluate model with this subset
            eval_result = evaluate_model(X_train, X_test, y_train, y_test, feature_indices)
            results.append(eval_result)
            
            # Track best model for this size
            if eval_result['test_r2'] > best_score:
                best_score = eval_result['test_r2']
                best_feature_set = feature_indices
        
        best_model_per_size.append({
            'n_features': k,
            'feature_indices': best_feature_set,
            'test_r2': best_score
        })
        
        print(f"Best model with {k} features has test R¬≤: {best_score:.4f}")
    
    print(f"Best Subset Selection completed in {time.time() - start_time:.2f} seconds")
    return results, best_model_per_size

# 5. Forward Stepwise Selection
def forward_stepwise_selection(X_train, X_test, y_train, y_test):
    n_features = X_train.shape[1]
    current_features = []
    remaining_features = list(range(n_features))
    results = []
    
    print("Running Forward Stepwise Selection...")
    start_time = time.time()
    
    for i in range(n_features):
        best_score = -np.inf
        best_feature = None
        
        # Try adding each remaining feature
        for feature in remaining_features:
            feature_set = current_features + [feature]
            eval_result = evaluate_model(X_train, X_test, y_train, y_test, feature_set)
            
            if eval_result['test_r2'] > best_score:
                best_score = eval_result['test_r2']
                best_feature = feature
                best_result = eval_result
        
        # Add the best feature
        if best_feature is not None:
            current_features.append(best_feature)
            remaining_features.remove(best_feature)
            results.append(best_result)
            print(f"Step {i+1}: Added feature {best_feature}, Test R¬≤: {best_score:.4f}")
        else:
            break
    
    print(f"Forward Stepwise Selection completed in {time.time() - start_time:.2f} seconds")
    return results

# 6. Backward Stepwise Selection
def backward_stepwise_selection(X_train, X_test, y_train, y_test):
    n_features = X_train.shape[1]
    current_features = list(range(n_features))
    results = []
    
    print("Running Backward Stepwise Selection...")
    start_time = time.time()
    
    # Evaluate model with all features
    full_model = evaluate_model(X_train, X_test, y_train, y_test, current_features)
    results.append(full_model)
    print(f"Full model with {n_features} features has Test R¬≤: {full_model['test_r2']:.4f}")
    
    for i in range(n_features - 1):
        best_score = -np.inf
        worst_feature = None
        
        # Try removing each feature
        for j, feature in enumerate(current_features):
            feature_set = current_features.copy()
            feature_set.pop(j)
            
            eval_result = evaluate_model(X_train, X_test, y_train, y_test, feature_set)
            
            if eval_result['test_r2'] > best_score:
                best_score = eval_result['test_r2']
                worst_feature = j
                best_result = eval_result
        
        # Remove the worst feature
        if worst_feature is not None:
            removed_feature = current_features.pop(worst_feature)
            results.append(best_result)
            print(f"Step {i+1}: Removed feature {removed_feature}, Test R¬≤: {best_score:.4f}")
        else:
            break
    
    print(f"Backward Stepwise Selection completed in {time.time() - start_time:.2f} seconds")
    return results

# Run all subset selection methods
# Note: We're using top features to make Best Subset computationally feasible
print("\n" + "="*50)
print("SUBSET SELECTION METHODS")
print("="*50)

# Using top features for these methods
print(f"Using top {len(top_features)} features for subset selection based on correlation")
print("Top features selected:", top_features)

# 4. Best Subset Selection (with limited features)
best_subset_results, best_models_by_size = best_subset_selection(
    X_train_top_scaled, X_test_top_scaled, y_train, y_test, max_features=10)

# 5. Forward Stepwise Selection
forward_results = forward_stepwise_selection(
    X_train_top_scaled, X_test_top_scaled, y_train, y_test)

# 6. Backward Stepwise Selection
backward_results = backward_stepwise_selection(
    X_train_top_scaled, X_test_top_scaled, y_train, y_test)

# Compare the results of all methods
def plot_subset_selection_comparison():
    plt.figure(figsize=(12, 8))
    
    # Extract data for plotting
    best_subset_data = [(model['n_features'], model['test_r2']) for model in best_models_by_size]
    best_subset_x, best_subset_y = zip(*best_subset_data)
    
    forward_data = [(result['n_features'], result['test_r2']) for result in forward_results]
    forward_x, forward_y = zip(*forward_data)
    
    backward_data = [(result['n_features'], result['test_r2']) for result in backward_results]
    backward_x, backward_y = zip(*backward_data)
    
    # Plot
    plt.plot(best_subset_x, best_subset_y, 'o-', label='Best Subset', linewidth=2)
    plt.plot(forward_x, forward_y, 's-', label='Forward Stepwise', linewidth=2)
    plt.plot(backward_x, backward_y, '^-', label='Backward Stepwise', linewidth=2)
    
    plt.xlabel('Number of Features', fontsize=14)
    plt.ylabel('Test R¬≤', fontsize=14)
    plt.title('Comparison of Subset Selection Methods', fontsize=16)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(fontsize=12)
    plt.tight_layout()
    plt.show()

plot_subset_selection_comparison()

# Identify optimal model size for each method
def get_optimal_model(results, criterion='test_r2'):
    # Find model with best test R¬≤
    best_idx = np.argmax([result[criterion] for result in results])
    return results[best_idx]

best_subset_optimal = max(best_models_by_size, key=lambda x: x['test_r2'])
forward_optimal = get_optimal_model(forward_results)
backward_optimal = get_optimal_model(backward_results)

print("\nOptimal Models Summary:")
print(f"Best Subset: {best_subset_optimal['n_features']} features, R¬≤ = {best_subset_optimal['test_r2']:.4f}")
print(f"Forward Stepwise: {forward_optimal['n_features']} features, R¬≤ = {forward_optimal['test_r2']:.4f}")
print(f"Backward Stepwise: {backward_optimal['n_features']} features, R¬≤ = {backward_optimal['test_r2']:.4f}")

# Extract and display the selected features for each optimal model
def get_feature_names(feature_indices, feature_names):
    return [feature_names[i] for i in feature_indices]

best_subset_features = get_feature_names(best_subset_optimal['feature_indices'], top_features)
forward_features = get_feature_names(forward_optimal['feature_indices'], top_features)
backward_features = get_feature_names(backward_optimal['feature_indices'], top_features)

print("\nSelected Features:")
print("Best Subset:", best_subset_features)
print("Forward Stepwise:", forward_features)
print("Backward Stepwise:", backward_features)

# Create a final comparison table of all methods (including shrinkage methods from previous analysis)
comparison_table = pd.DataFrame({
    'Method': ['Ridge', 'LASSO', 'Elastic Net', 'Best Subset', 'Forward Stepwise', 'Backward Stepwise'],
    'Test R¬≤': [
        results_df.loc[0, 'Test R¬≤'],  # Ridge
        results_df.loc[1, 'Test R¬≤'],  # LASSO
        results_df.loc[2, 'Test R¬≤'],  # Elastic Net
        best_subset_optimal['test_r2'], 
        forward_optimal['test_r2'], 
        backward_optimal['test_r2']
    ],
    'Number of Features': [
        results_df.loc[0, 'Features Selected'],  # Ridge
        results_df.loc[1, 'Features Selected'],  # LASSO
        results_df.loc[2, 'Features Selected'],  # Elastic Net
        best_subset_optimal['n_features'],
        forward_optimal['n_features'],
        backward_optimal['n_features']
    ]
})

# Sort by performance
comparison_table = comparison_table.sort_values('Test R¬≤', ascending=False)

print("\nFinal Comparison of All Methods:")
print(comparison_table)

# Create a bar chart to compare all methods
plt.figure(figsize=(12, 6))
plt.bar(comparison_table['Method'], comparison_table['Test R¬≤'])
plt.xlabel('Method')
plt.ylabel('Test R¬≤')
plt.title('Performance Comparison of All Methods')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Analyze feature consistency across methods
def analyze_feature_consistency():
    # Collect selected features from all methods
    all_selected_features = {
        'LASSO': lasso_coef[lasso_coef != 0].index.tolist(),
        'Elastic Net': enet_coef[enet_coef != 0].index.tolist(),
        'Best Subset': best_subset_features,
        'Forward Stepwise': forward_features,
        'Backward Stepwise': backward_features
    }
    
    # Count feature occurrences
    feature_count = {}
    for method, features in all_selected_features.items():
        for feature in features:
            if feature in feature_count:
                feature_count[feature] += 1
            else:
                feature_count[feature] = 1
    
    # Create and sort dataframe
    consistency_df = pd.DataFrame({
        'Feature': list(feature_count.keys()),
        'Count': list(feature_count.values())
    })
    consistency_df = consistency_df.sort_values('Count', ascending=False)
    
    # Display top 10 most consistent features
    print("\nTop 10 Most Consistent Features Across Methods:")
    print(consistency_df.head(10))
    
    # Plot
    plt.figure(figsize=(12, 6))
    plt.bar(consistency_df['Feature'].head(10), consistency_df['Count'].head(10))
    plt.xlabel('Feature')
    plt.ylabel('Number of Methods Including Feature')
    plt.title('Feature Consistency Across Different Selection Methods')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
    
    return consistency_df

feature_consistency = analyze_feature_consistency()

# Summary and conclusion
print("\nCONCLUSION AND RECOMMENDATIONS")
print("="*50)
print(f"Best Overall Method: {comparison_table.iloc[0]['Method']} (R¬≤ = {comparison_table.iloc[0]['Test R¬≤']:.4f})")
print("\nMost Important Features for Predicting Quality of Sleep:")
for feature in feature_consistency['Feature'].head(5):
    print(f"- {feature}")
````
