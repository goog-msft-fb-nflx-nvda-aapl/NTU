(10%) Prove that the determinant of the sample covariance matrix ùêí satisfies the equation  $|ùêí| = |ùêë| * \prod_{i=1,2,...p}{ùë†_{ùëñùëñ}} $ where ùêë is the corresponding sample correlation matrix.

### **2-1: Definitions**  [üö®SlideDA01 Page36]
Let $ \mathbf{S} $ be the $ p \times p $ sample covariance matrix, whose elements are given by:
- $S_{ij} = \text{Cov}(X_i, X_j)$
- $ S_{ii} = \text{Var}(X_i) = \sigma_i^2 $.
- $\sigma_i = \sqrt{s_{ii}}$

The corresponding **sample correlation matrix** $ \mathbf{R} $ is defined as:
 - $R_{ij} = \frac{S_{ij}}{\sqrt{S_{ii} S_{jj}}} = \frac{S_{ij}}{\sigma_i \sigma_j}$
which normalizes the covariance values.
 - $S_{ij} = \sigma_i R_{ij} \sigma_j$

### **2-2: Expressing $ \mathbf{S} $ in Terms of $ \mathbf{R} $**
We can write $ \mathbf{S} $ in matrix form as:
 - $\mathbf{S} = \mathbf{D} \mathbf{R} \mathbf{D}$
where $ \mathbf{D} $ is a diagonal matrix containing the standard deviations $ \sqrt{S_{ii}} $:

 - $D = \text{diag}(\sigma_1, \sigma_2, ..., \sigma_p)$   

$
\mathbf{D} = \begin{bmatrix} 
\sqrt{S_{11}} & 0 & \dots & 0 \\ 
0 & \sqrt{S_{22}} & \dots & 0 \\ 
\vdots & \vdots & \ddots & \vdots \\ 
0 & 0 & \dots & \sqrt{S_{pp}}
\end{bmatrix}
$

Since $ S_{ij} = R_{ij} \sqrt{S_{ii} S_{jj}} $, the covariance matrix can be rewritten as:
$
\mathbf{S} = \mathbf{D} \mathbf{R} \mathbf{D}
$

### **2-3: Compute Determinant of $ \mathbf{S} $**
Using the determinant property:

$
\det(\mathbf{S}) = \det(\mathbf{D} \mathbf{R} \mathbf{D})
$

Since the determinant of a product satisfies:

$
\det(\mathbf{S}) = \det(\mathbf{D}) \det(\mathbf{R}) \det(\mathbf{D})
$

And since $ \det(\mathbf{D}) $ is the product of the diagonal elements:

$
\det(\mathbf{D}) = \prod_{i=1}^{p} \sqrt{S_{ii}}
$

Thus:

$
\det(\mathbf{S}) = (\prod_{i=1}^{p} \sqrt{S_{ii}}) \det(\mathbf{R}) (\prod_{i=1}^{p} \sqrt{S_{ii}})
$

$
= \det(\mathbf{R}) \cdot \prod_{i=1}^{p} S_{ii}
$

which proves the given formula:

$
|\mathbf{S}| = |\mathbf{R}| \times \prod_{i=1}^{p} S_{ii}
$

### üö® [Slide DA03 Page16 ]   
Hotelling‚Äôs $ T^2 $ statistic is used to test whether the mean vector of a sample differs significantly from a specified population mean vector. The formula for $ T^2 $ is:

$
T^2 = n (\bar{X} - \mu)^T S^{-1} (\bar{X} - \mu)
$

where:
- $ n $ is the sample size,
- $ \bar{X} $ is the sample mean vector,
- $ \mu $ is the hypothesized mean vector,
- $ S $ is the sample covariance matrix.

### 4-a-1: Calculate the sample mean vector $\bar{X}$

From the data:


Sample mean xÃÑ = [6, 10]


### 4-a-2: Calculate the sample covariance matrix
$
S = \frac{1}{n-1} \sum_{i=1}^{n} (X_i - \bar{X})(X_i - \bar{X})^T
$

### 4-a-5: Calculate Hotelling's T¬≤ statistic
$
T^2 = n (\bar{X} - \mu)^T S^{-1} (\bar{X} - \mu)
$  
 - T¬≤ = n(xÃÑ - Œº‚ÇÄ)·µÄS‚Åª¬π(xÃÑ - Œº‚ÇÄ)
 - T¬≤ = 4 √ó [-1, -1] √ó [[0.409, 0.682], [0.682, 1.636]] √ó [-1, -1]·µÄ
 - T¬≤ = 13.64

$\Rightarrow$ Hotelling's T¬≤ statistic for testing H‚ÇÄ: Œº·µÄ = [7, 11] is 13.64.

We will validate this using Python.

````python
import numpy as np
from scipy import stats

# Given data
X = np.array([
    [2, 12],
    [8, 9],
    [6, 9],
    [8, 10]
])

# Hypothesized mean
mu0 = np.array([7, 11])

# Step 1: Calculate sample size, dimension, and sample mean
n, p = X.shape
xbar = np.mean(X, axis=0)
print(f"Sample mean: {xbar}")

# Step 2: Calculate the sample covariance matrix
S = np.cov(X, rowvar=False)
print(f"Sample covariance matrix S:\n{S}")

# Step 3: Calculate the difference betWeen sample mean and hypothesized mean
diff = xbar - mu0
print(f"Mean difference (xbar - mu0): {diff}")

# Step 4: Calculate the inverse of the covariance matrix
S_inv = np.linalg.inv(S)
print(f"Inverse of covariance matrix S^-1:\n{S_inv}")

# Step 5: Calculate Hotelling's T^2 statistic
T2 = n * diff.T @ S_inv @ diff
print(f"Hotelling's T^2 statistic: {T2}")

# Optional: Calculate the equivalent F-statistic for testing
F = (n - p) / (p * (n - 1)) * T2
F_critical = stats.f.ppf(0.95, p, n - p)
p_value = 1 - stats.f.cdf(F, p, n - p)

print(f"Equivalent F-statistic: {F}")
print(f"Critical F-value (Œ±=0.05): {F_critical}")
print(f"p-value: {p_value}")
````

### Question 4-b. Distribution of the T¬≤ Statistic

Hotelling's T¬≤ statistic follows a scaled F-distribution under the null hypothesis. Specifically:

$\frac{n-p}{p(n-1)}T^2 \sim F_{p,n-p}$

Where:
- n = 4 (sample size)
- p = 2 (dimension of the data)
- T¬≤ = 13.64 (the calculated T¬≤ value)

In this case, the distribution is:

$\frac{4-2}{2(4-1)}T^2 = \frac{2}{6}T^2 = \frac{1}{3}T^2 \sim F_{2,2}$  
$T^2 \sim 3 \times F_{2,2}$

So the T¬≤ statistic (scaled by 1/3) follows an F-distribution with 2 numerator degrees of freedom and 2 denominator degrees of freedom.
**Hotelling‚Äôs $ T^2(2, 4) $ distribution**, which is equivalent to an **F(2,2) distribution** when transformed using the formula above.

The calculated F value is approximately 4.55, which is derived from:
$F = \frac{n-p}{p(n-1)}T^2 = \frac{2}{6} \times 13.64 = 4.55$

This F-value can be compared against the critical value of the F-distribution with (p, n-p) = (2, 2) degrees of freedom to determine statistical significance at a chosen alpha level.

   - Null Hypothesis ($ H_0 $): The mean vectors of male and female turtles are equal ($ \mu_{\text{female}} = \mu_{\text{male}} $).
   - Alternative Hypothesis ($ H_1 $): The mean vectors of male and female turtles are not equal ($ \mu_{\text{female}} \neq \mu_{\text{male}} $).

3. **Calculate Mean Vectors**:
   Compute the mean vectors for females and males for the log-transformed data.

4. **Compute Pooled Covariance Matrix**:
   Combine the covariance matrices of male and female groups into a pooled covariance matrix.

5. **Hotelling's $ T^2 $ Test Statistic**:
   Use the formula:
   $
   T^2 = n_1 n_2 / (n_1 + n_2) (\bar{X}_1 - \bar{X}_2)^T S^{-1} (\bar{X}_1 - \bar{X}_2)
   $
   where $ S^{-1} $ is the inverse of the pooled covariance matrix, $ n_1 = n_2 = 24 $, and $ (\bar{X}_1 - \bar{X}_2) $ is the difference in mean vectors.

6. **Convert to F-Statistic**:
   Convert $ T^2 $ to an F-statistic for comparison with critical values:
   $
   F = \frac{(n_1 + n_2 - p - 1)}{p (n_1 + n_2 - 2)} T^2
   $
   where $ p = 3 $ (number of variables).

7. **Decision Rule**:
   Compare the computed F-statistic with the critical value from the F-distribution at $ \alpha = 0.05 $. Reject $ H_0 $ if $ F > F_{\text{critical}}(p, n_1 + n_2 - p - 1) $.

````python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
from scipy.linalg import inv

# Convert to DataFrame for easier manipulation
df = pd.DataFrame(data, columns=header)

# Extract female and male data
female_data = df[["FemaleLength", "FemaleWidth", "FemaleHeight"]].values
male_data = df[["MaleLength", "MaleWidth", "MaleHeight"]].values

# Apply log transformation
log_female_data = np.log(female_data)
log_male_data = np.log(male_data)

# Calculate sample sizes and dimension
n1 = female_data.shape[0]  # Female sample size
n2 = male_data.shape[0]    # Male sample size
p = female_data.shape[1]   # Dimension (number of variables)

# Calculate sample means
female_means = np.mean(female_data, axis=0)
male_means = np.mean(male_data, axis=0)
log_female_means = np.mean(log_female_data, axis=0)
log_male_means = np.mean(log_male_data, axis=0)

print("Female Means (original):", female_means)
print("Male Means (original):", male_means)
print("Female Means (log-transformed):", log_female_means)
print("Male Means (log-transformed):", log_male_means)

# Calculate covariance matrices
female_cov = np.cov(log_female_data, rowvar=False)
male_cov = np.cov(log_male_data, rowvar=False)

# Calculate pooled covariance matrix
pooled_cov = ((n1 - 1) * female_cov + (n2 - 1) * male_cov) / (n1 + n2 - 2)
print("\nPooled Covariance Matrix (log-transformed):")
print(pooled_cov)

# Calculate Hotelling's T^2 statistic
mean_diff = log_female_means - log_male_means
inv_pooled_cov = inv(pooled_cov)
T2 = (n1 * n2) / (n1 + n2) * mean_diff.dot(inv_pooled_cov).dot(mean_diff)
print("\nHotelling's T^2 statistic:", T2)

# Convert T^2 to F-statistic
F_stat = ((n1 + n2 - p - 1) / ((n1 + n2 - 2) * p)) * T2
print("F-statistic:", F_stat)

# Calculate degrees of freedom
df1 = p
df2 = n1 + n2 - p - 1
print("Degrees of freedom:", df1, df2)

# Calculate critical F-value at Œ± = 0.05
critical_F = stats.f.ppf(1 - 0.05, df1, df2)
print("Critical F-value (Œ± = 0.05):", critical_F)

# Calculate p-value
p_value = 1 - stats.f.cdf(F_stat, df1, df2)
print("p-value:", p_value)

# Decision
alpha = 0.05
if F_stat > critical_F:
    print("\nDecision: Reject H0. The mean vectors are significantly different (p < {}).".format(alpha))
else:
    print("\nDecision: Fail to reject H0. There is not enough evidence to conclude the mean vectors are different.")

# Calculate Mahalanobis distance betWeen the means
mahalanobis_dist = np.sqrt(mean_diff.dot(inv_pooled_cov).dot(mean_diff))
print("Mahalanobis distance betWeen mean vectors:", mahalanobis_dist)

# Visualize the differences
plt.figure(figsize=(15, 10))

# Original data
plt.subplot(2, 3, 1)
plt.scatter(female_data[:, 0], female_data[:, 1], label='Female', color='red', alpha=0.7)
plt.scatter(male_data[:, 0], male_data[:, 1], label='Male', color='blue', alpha=0.7)
plt.xlabel('Length')
plt.ylabel('Width')
plt.title('Original Data: Length vs Width')
plt.legend()

plt.subplot(2, 3, 2)
plt.scatter(female_data[:, 0], female_data[:, 2], label='Female', color='red', alpha=0.7)
plt.scatter(male_data[:, 0], male_data[:, 2], label='Male', color='blue', alpha=0.7)
plt.xlabel('Length')
plt.ylabel('Height')
plt.title('Original Data: Length vs Height')
plt.legend()

plt.subplot(2, 3, 3)
plt.scatter(female_data[:, 1], female_data[:, 2], label='Female', color='red', alpha=0.7)
plt.scatter(male_data[:, 1], male_data[:, 2], label='Male', color='blue', alpha=0.7)
plt.xlabel('Width')
plt.ylabel('Height')
plt.title('Original Data: Width vs Height')
plt.legend()

# Log-transformed data
plt.subplot(2, 3, 4)
plt.scatter(log_female_data[:, 0], log_female_data[:, 1], label='Female', color='red', alpha=0.7)
plt.scatter(log_male_data[:, 0], log_male_data[:, 1], label='Male', color='blue', alpha=0.7)
plt.xlabel('Log(Length)')
plt.ylabel('Log(Width)')
plt.title('Log-transformed Data: Length vs Width')
plt.legend()

plt.subplot(2, 3, 5)
plt.scatter(log_female_data[:, 0], log_female_data[:, 2], label='Female', color='red', alpha=0.7)
plt.scatter(log_male_data[:, 0], log_male_data[:, 2], label='Male', color='blue', alpha=0.7)
plt.xlabel('Log(Length)')
plt.ylabel('Log(Height)')
plt.title('Log-transformed Data: Length vs Height')
plt.legend()

plt.subplot(2, 3, 6)
plt.scatter(log_female_data[:, 1], log_female_data[:, 2], label='Female', color='red', alpha=0.7)
plt.scatter(log_male_data[:, 1], log_male_data[:, 2], label='Male', color='blue', alpha=0.7)
plt.xlabel('Log(Width)')
plt.ylabel('Log(Height)')
plt.title('Log-transformed Data: Width vs Height')
plt.legend()

plt.tight_layout()
plt.savefig('turtle_comparison.png')
plt.show()

# Additional: Univariate t-tests for comparison
print("\nUnivariate t-tests for each dimension (log-transformed):")
for i, dim in enumerate(['Length', 'Width', 'Height']):
    t_stat, p_val = stats.ttest_ind(log_female_data[:, i], log_male_data[:, i], equal_var=True)
    print(f"{dim}: t = {t_stat:.4f}, p = {p_val:.6f}")
````

As noted, with larger sample sizes, these statistics would converge. In this case, the low sample size causes the pillai's trace to indicate a possible species effect, while the Wilks' Lambda did not.
* **Univariate vs. Multivariate Analysis:** ANOVA examines each dependent variable separately, potentially missing interactive effects or patterns across multiple variables. MANOVA, on the other hand, considers the combined effect on all dependent variables, potentially revealing subtler relationships that univariate analysis might overlook. The individual ANOVA was able to detect the strong species effect in the 720CM data, while the MANOVA was not able to detect it as strongly.
