## üìí Appendix 1
Code and Visualization for Problem 1.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.pipeline import Pipeline
import warnings
warnings.filterwarnings('ignore')

# Set the plotting style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)

# Load the dataset
df = pd.read_csv('/Users/chun-feitan/Desktop/Sleep_health_and_lifestyle_dataset.csv')

# Data Preprocessing
# Replace missing values in the 'Sleep Disorder' column with 'None'
df['Sleep Disorder'] = df['Sleep Disorder'].fillna('None')
# Drop duplicates
df = df.drop_duplicates()

# Display basic information about the dataset
print(f"Dataset shape after removing duplicates: {df.shape}")
print("\nFirst few rows of the dataset:")
print(df.head())

print("\nBasic statistics:")
print(df.describe())

print("\nMissing values in each column:")
print(df.isnull().sum())

# Exploring the target variable distribution
sleep_disorder_counts = df['Sleep Disorder'].value_counts()
print("\nSleep Disorder distribution:")
print(sleep_disorder_counts)

# Visualize the distribution of sleep disorders
plt.figure(figsize=(10, 6))
sns.countplot(x='Sleep Disorder', data=df, palette='viridis')
plt.title('Distribution of Sleep Disorders', fontsize=16)
plt.xlabel('Sleep Disorder', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.tight_layout()
plt.savefig('sleep_disorder_distribution.png')

# Feature engineering
# Extract systolic and diastolic BP from the Blood Pressure column
df[['Systolic BP', 'Diastolic BP']] = df['Blood Pressure'].str.split('/', expand=True).astype(int)

# Encoding categorical variables
le_gender = LabelEncoder()
le_bmi = LabelEncoder()
le_occupation = LabelEncoder()

df['Gender_encoded'] = le_gender.fit_transform(df['Gender'])
df['BMI_encoded'] = le_bmi.fit_transform(df['BMI Category'])
df['Occupation_encoded'] = le_occupation.fit_transform(df['Occupation'])

# Create a target variable encoder
le_target = LabelEncoder()
df['Sleep Disorder_encoded'] = le_target.fit_transform(df['Sleep Disorder'])

# Map encoded values back to original labels for reference
print("\nSleep Disorder Encoding:")
for i, label in enumerate(le_target.classes_):
    print(f"{label}: {i}")

# Create correlation heatmap to identify important predictors
numeric_cols = ['Age', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 
                'Stress Level', 'Heart Rate', 'Daily Steps', 'Systolic BP', 'Diastolic BP', 
                'Gender_encoded', 'BMI_encoded', 'Occupation_encoded', 'Sleep Disorder_encoded']

plt.figure(figsize=(14, 10))
correlation_matrix = df[numeric_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f')
plt.title('Correlation Matrix', fontsize=16)
plt.tight_layout()
plt.savefig('correlation_matrix.png')

# Pairplot for key features vs sleep disorder
key_features = ['Age', 'Sleep Duration', 'Quality of Sleep', 'Stress Level', 'BMI_encoded']
plt.figure(figsize=(16, 12))
sns.pairplot(df, vars=key_features, hue='Sleep Disorder', palette='viridis')
plt.tight_layout()
plt.savefig('sleep_features_pairplot.png')

# Define features for modeling
features = ['Age', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 
            'Stress Level', 'Heart Rate', 'Daily Steps', 'Systolic BP', 'Diastolic BP',
            'Gender_encoded', 'BMI_encoded', 'Occupation_encoded']

X = df[features]
y = df['Sleep Disorder_encoded']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Create a figure to demonstrate class distributions
plt.figure(figsize=(10, 6))
train_dist = pd.Series(y_train).value_counts()
test_dist = pd.Series(y_test).value_counts()

# Calculate percentages
train_pct = train_dist / train_dist.sum() * 100
test_pct = test_dist / test_dist.sum() * 100

# Create a DataFrame for better visualization
dist_df = pd.DataFrame({
    'Training Set': train_pct,
    'Testing Set': test_pct
})

# Map indices back to original labels
dist_df.index = [le_target.inverse_transform([i])[0] for i in dist_df.index]

dist_df.plot(kind='bar')
plt.title('Class Distribution in Training and Testing Sets (%)', fontsize=16)
plt.ylabel('Percentage', fontsize=12)
plt.xlabel('Sleep Disorder', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('class_distribution.png')

# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\n--- Method 1: Discriminant Analysis ---")

# Linear Discriminant Analysis (LDA)
lda = LinearDiscriminantAnalysis()
lda.fit(X_train_scaled, y_train)
y_pred_lda = lda.predict(X_test_scaled)

print("\nLinear Discriminant Analysis Results:")
print("Accuracy:", accuracy_score(y_test, y_pred_lda))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_lda, target_names=le_target.classes_))

# Confusion Matrix for LDA
plt.figure(figsize=(8, 6))
cm_lda = confusion_matrix(y_test, y_pred_lda)
sns.heatmap(cm_lda, annot=True, fmt='d', cmap='Blues', xticklabels=le_target.classes_, yticklabels=le_target.classes_)
plt.title('Confusion Matrix - Linear Discriminant Analysis', fontsize=16)
plt.ylabel('True Label', fontsize=12)
plt.xlabel('Predicted Label', fontsize=12)
plt.tight_layout()
plt.savefig('lda_confusion_matrix.png')

# Quadratic Discriminant Analysis (QDA)
qda = QuadraticDiscriminantAnalysis()
qda.fit(X_train_scaled, y_train)
y_pred_qda = qda.predict(X_test_scaled)

print("\nQuadratic Discriminant Analysis Results:")
print("Accuracy:", accuracy_score(y_test, y_pred_qda))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_qda, target_names=le_target.classes_))

# Confusion Matrix for QDA
plt.figure(figsize=(8, 6))
cm_qda = confusion_matrix(y_test, y_pred_qda)
sns.heatmap(cm_qda, annot=True, fmt='d', cmap='Blues', xticklabels=le_target.classes_, yticklabels=le_target.classes_)
plt.title('Confusion Matrix - Quadratic Discriminant Analysis', fontsize=16)
plt.ylabel('True Label', fontsize=12)
plt.xlabel('Predicted Label', fontsize=12)
plt.tight_layout()
plt.savefig('qda_confusion_matrix.png')

# Cross Validation for LDA and QDA
cv_lda = cross_val_score(lda, X_train_scaled, y_train, cv=5)
cv_qda = cross_val_score(qda, X_train_scaled, y_train, cv=5)

print("\nLDA Cross-Validation Scores:", cv_lda)
print("LDA Mean CV Score:", cv_lda.mean())
print("\nQDA Cross-Validation Scores:", cv_qda)
print("QDA Mean CV Score:", cv_qda.mean())

print("\n--- Method 2: Support Vector Machine ---")

# SVM with default parameters
svm = SVC()
svm.fit(X_train_scaled, y_train)
y_pred_svm = svm.predict(X_test_scaled)

print("\nSupport Vector Machine Results (Default Parameters):")
print("Accuracy:", accuracy_score(y_test, y_pred_svm))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_svm, target_names=le_target.classes_))

# Confusion Matrix for SVM
plt.figure(figsize=(8, 6))
cm_svm = confusion_matrix(y_test, y_pred_svm)
sns.heatmap(cm_svm, annot=True, fmt='d', cmap='Blues', xticklabels=le_target.classes_, yticklabels=le_target.classes_)
plt.title('Confusion Matrix - Support Vector Machine (Default)', fontsize=16)
plt.ylabel('True Label', fontsize=12)
plt.xlabel('Predicted Label', fontsize=12)
plt.tight_layout()
plt.savefig('svm_default_confusion_matrix.png')

# Grid search for SVM parameters
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf', 'linear', 'poly']
}

grid_search = GridSearchCV(SVC(), param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train_scaled, y_train)

print("\nBest SVM Parameters:", grid_search.best_params_)
best_svm = grid_search.best_estimator_
y_pred_best_svm = best_svm.predict(X_test_scaled)

print("\nSupport Vector Machine Results (Optimized Parameters):")
print("Accuracy:", accuracy_score(y_test, y_pred_best_svm))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_best_svm, target_names=le_target.classes_))

# Confusion Matrix for optimized SVM
plt.figure(figsize=(8, 6))
cm_best_svm = confusion_matrix(y_test, y_pred_best_svm)
sns.heatmap(cm_best_svm, annot=True, fmt='d', cmap='Blues', xticklabels=le_target.classes_, yticklabels=le_target.classes_)
plt.title('Confusion Matrix - Support Vector Machine (Optimized)', fontsize=16)
plt.ylabel('True Label', fontsize=12)
plt.xlabel('Predicted Label', fontsize=12)
plt.tight_layout()
plt.savefig('svm_optimized_confusion_matrix.png')

# Cross Validation for SVM
cv_svm_default = cross_val_score(SVC(), X_train_scaled, y_train, cv=5)
cv_svm_best = cross_val_score(best_svm, X_train_scaled, y_train, cv=5)

print("\nDefault SVM Cross-Validation Scores:", cv_svm_default)
print("Default SVM Mean CV Score:", cv_svm_default.mean())
print("\nOptimized SVM Cross-Validation Scores:", cv_svm_best)
print("Optimized SVM Mean CV Score:", cv_svm_best.mean())

# Feature importance for LDA
plt.figure(figsize=(12, 6))
importance = lda.coef_
feature_importance = pd.DataFrame(importance, columns=features)
feature_importance_abs = feature_importance.abs().mean(axis=0).sort_values(ascending=False)
sns.barplot(x=feature_importance_abs.values, y=feature_importance_abs.index, palette='viridis')
plt.title('Feature Importance in Linear Discriminant Analysis', fontsize=16)
plt.xlabel('Absolute Coefficient Magnitude', fontsize=12)
plt.tight_layout()
plt.savefig('lda_feature_importance.png')

# Summarize the results
models = ['LDA', 'QDA', 'SVM (Default)', 'SVM (Optimized)']
test_accuracies = [
    accuracy_score(y_test, y_pred_lda),
    accuracy_score(y_test, y_pred_qda),
    accuracy_score(y_test, y_pred_svm),
    accuracy_score(y_test, y_pred_best_svm)
]
cv_scores = [cv_lda.mean(), cv_qda.mean(), cv_svm_default.mean(), cv_svm_best.mean()]

results_df = pd.DataFrame({
    'Model': models,
    'Test Accuracy': test_accuracies,
    'CV Accuracy': cv_scores
})

print("\nModel Performance Comparison:")
print(results_df)

plt.figure(figsize=(12, 6))
results_df_melted = pd.melt(results_df, id_vars='Model', var_name='Metric', value_name='Accuracy')
sns.barplot(x='Model', y='Accuracy', hue='Metric', data=results_df_melted, palette='viridis')
plt.title('Model Performance Comparison', fontsize=16)
plt.xlabel('Model', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.ylim(0, 1)
plt.xticks(rotation=0)
plt.legend(title='Metric')
plt.tight_layout()
plt.savefig('model_comparison.png')

# Print the classification reports in a formatted way
print("\n--- Detailed Classification Reports ---")
models_preds = {
    'LDA': y_pred_lda,
    'QDA': y_pred_qda,
    'SVM (Default)': y_pred_svm,
    'SVM (Optimized)': y_pred_best_svm
}

for model_name, y_pred in models_preds.items():
    print(f"\n{model_name} Classification Report:")
    print(classification_report(y_test, y_pred, target_names=le_target.classes_))
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.svm import SVR, SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, mean_squared_error, r2_score
from sklearn.multiclass import OneVsRestClassifier
from sklearn.multioutput import MultiOutputRegressor
import warnings

warnings.filterwarnings('ignore')

# Load the dataset
try:
    df = pd.read_csv('/Users/chun-feitan/Desktop/Sleep_health_and_lifestyle_dataset.csv')
    print("Dataset loaded successfully.")
except FileNotFoundError:
    print("File not found. Please update the file path.")
    raise

# Data Cleaning
print("\n--- Data Cleaning ---")
# Fill missing values in 'Sleep Disorder' column with 'None'
df['Sleep Disorder'] = df['Sleep Disorder'].fillna('None')
# Remove duplicates
df_clean = df.drop_duplicates().reset_index(drop=True)
print(f"Original shape: {df.shape}, After removing duplicates: {df_clean.shape}")

# Display basic information about the dataset
print("\n--- Dataset Info ---")
print(f"Number of records: {len(df_clean)}")
print(f"Number of features: {len(df_clean.columns)}")
print(f"Features: {', '.join(df_clean.columns)}")

# Data Exploration
print("\n--- Data Exploration ---")

# Check data types
print("\nData types:")
print(df_clean.dtypes)

# Display summary statistics
print("\nSummary statistics:")
print(df_clean.describe().T)

# Check for missing values
print("\nMissing values:")
print(df_clean.isnull().sum())

# Display distribution of the target variable
print("\nSleep Disorder distribution:")
sleep_disorder_counts = df_clean['Sleep Disorder'].value_counts()
print(sleep_disorder_counts)
print(f"Sleep Disorder proportions:\n{sleep_disorder_counts / len(df_clean)}")

# Process Blood Pressure column
print("\n--- Processing Blood Pressure ---")
# Split Blood Pressure into systolic and diastolic
df_clean[['Systolic', 'Diastolic']] = df_clean['Blood Pressure'].str.split('/', expand=True).astype(int)
print("Blood Pressure split into Systolic and Diastolic columns.")

# Feature Engineering
print("\n--- Feature Engineering ---")

# Create BMI numeric values based on categories
bmi_mapping = {
    'Normal': 2,
    'Normal Weight': 2,
    'Overweight': 3,
    'Obese': 4
}
df_clean['BMI_Numeric'] = df_clean['BMI Category'].map(bmi_mapping)
print("BMI categories mapped to numeric values.")

# Encode target variable
label_encoder = LabelEncoder()
df_clean['Sleep_Disorder_Encoded'] = label_encoder.fit_transform(df_clean['Sleep Disorder'])
print("Sleep disorders encoded:", dict(zip(label_encoder.classes_, range(len(label_encoder.classes_)))))

# Define features and target
X = df_clean.drop(['Person ID', 'Sleep Disorder', 'Sleep_Disorder_Encoded', 'Blood Pressure'], axis=1)
y = df_clean['Sleep_Disorder_Encoded']

# Identify categorical and numerical columns
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()

print(f"\nCategorical features: {categorical_cols}")
print(f"Numerical features: {numerical_cols}")

# Set up the preprocessing for the pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
print(f"\nTraining set size: {X_train.shape[0]}, Test set size: {X_test.shape[0]}")

# Data Visualization
print("\n--- Data Visualization ---")

# Create a figure with subplots
plt.figure(figsize=(20, 15))

# 1. Sleep Disorder Distribution
plt.subplot(3, 3, 1)
sns.countplot(x='Sleep Disorder', data=df_clean)
plt.title('Distribution of Sleep Disorders')
plt.xticks(rotation=45)

# 2. Sleep Duration vs Quality of Sleep by Sleep Disorder
plt.subplot(3, 3, 2)
sns.scatterplot(x='Sleep Duration', y='Quality of Sleep', hue='Sleep Disorder', data=df_clean)
plt.title('Sleep Duration vs Quality of Sleep')

# 3. Age Distribution by Sleep Disorder
plt.subplot(3, 3, 3)
sns.boxplot(x='Sleep Disorder', y='Age', data=df_clean)
plt.title('Age Distribution by Sleep Disorder')
plt.xticks(rotation=45)

# 4. Stress Level Distribution by Sleep Disorder
plt.subplot(3, 3, 4)
sns.boxplot(x='Sleep Disorder', y='Stress Level', data=df_clean)
plt.title('Stress Level Distribution by Sleep Disorder')
plt.xticks(rotation=45)

# 5. Physical Activity Distribution by Sleep Disorder
plt.subplot(3, 3, 5)
sns.boxplot(x='Sleep Disorder', y='Physical Activity Level', data=df_clean)
plt.title('Physical Activity Distribution by Sleep Disorder')
plt.xticks(rotation=45)

# 6. Heart Rate Distribution by Sleep Disorder
plt.subplot(3, 3, 6)
sns.boxplot(x='Sleep Disorder', y='Heart Rate', data=df_clean)
plt.title('Heart Rate Distribution by Sleep Disorder')
plt.xticks(rotation=45)

# 7. BMI Category Distribution by Sleep Disorder
plt.subplot(3, 3, 7)
sns.countplot(x='BMI Category', hue='Sleep Disorder', data=df_clean)
plt.title('BMI Category Distribution by Sleep Disorder')
plt.xticks(rotation=45)

# 8. Gender Distribution by Sleep Disorder
plt.subplot(3, 3, 8)
sns.countplot(x='Gender', hue='Sleep Disorder', data=df_clean)
plt.title('Gender Distribution by Sleep Disorder')

# 9. Blood Pressure (Systolic) by Sleep Disorder
plt.subplot(3, 3, 9)
sns.boxplot(x='Sleep Disorder', y='Systolic', data=df_clean)
plt.title('Systolic Blood Pressure by Sleep Disorder')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()  # Display instead of saving

print("Visualizations displayed in notebook")

# Calculate correlation matrix for numeric variables
correlation_matrix = df_clean[numerical_cols + ['Sleep_Disorder_Encoded']].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()  # Display instead of saving

print("Correlation matrix displayed in notebook")

# Feature Importance Analysis
print("\n--- Feature Importance Analysis ---")

# Process the data with the preprocessor
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

# Get feature names after preprocessing
feature_names = (
    numerical_cols +
    list(preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_cols))
)

# Method 1: Using Linear Regression for feature importance
print("\n1. Linear Regression for Sleep Disorder Prediction")

# Linear Regression Model
lr_model = LinearRegression()
lr_model.fit(X_train_processed, y_train)

# Get feature importance from linear regression
if hasattr(lr_model, 'coef_') and lr_model.coef_.shape[0] == len(feature_names):
    lr_importance = pd.DataFrame({
        'Feature': feature_names,
        'Importance': np.abs(lr_model.coef_)
    }).sort_values(by='Importance', ascending=False)
    print("\nLinear Regression Feature Importance:")
    print(lr_importance.head(10))

# Predict on test set
lr_pred = lr_model.predict(X_test_processed)
# Round predictions to nearest integer for classification
lr_pred_class = np.round(lr_pred).astype(int)
# Clip predictions to be within valid range
lr_pred_class = np.clip(lr_pred_class, 0, len(label_encoder.classes_) - 1)

# Evaluate Linear Regression Model
print("\nLinear Regression Performance Metrics:")
print(f"Mean Squared Error: {mean_squared_error(y_test, lr_pred):.4f}")
print(f"R¬≤ Score: {r2_score(y_test, lr_pred):.4f}")
print(f"Accuracy (based on rounded predictions): {accuracy_score(y_test, lr_pred_class):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, lr_pred_class, target_names=label_encoder.classes_))

# Confusion Matrix for Linear Regression
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, lr_pred_class)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix - Linear Regression')
plt.tight_layout()
plt.show()  # Display instead of saving

print("Linear Regression confusion matrix displayed in notebook")

# Method 2: Using Support Vector Regression
print("\n2. Support Vector Regression for Sleep Disorder Prediction")

# SVR Model
svr_model = SVR(kernel='rbf', C=1.0, gamma='scale')
svr_model.fit(X_train_processed, y_train)

# Predict on test set
svr_pred = svr_model.predict(X_test_processed)
# Round predictions to nearest integer for classification
svr_pred_class = np.round(svr_pred).astype(int)
# Clip predictions to be within valid range
svr_pred_class = np.clip(svr_pred_class, 0, len(label_encoder.classes_) - 1)

# Evaluate SVR Model
print("\nSupport Vector Regression Performance Metrics:")
print(f"Mean Squared Error: {mean_squared_error(y_test, svr_pred):.4f}")
print(f"R¬≤ Score: {r2_score(y_test, svr_pred):.4f}")
print(f"Accuracy (based on rounded predictions): {accuracy_score(y_test, svr_pred_class):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, svr_pred_class, target_names=label_encoder.classes_))

# Confusion Matrix for SVR
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, svr_pred_class)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix - Support Vector Regression')
plt.tight_layout()
plt.show()  # Display instead of saving

print("SVR confusion matrix displayed in notebook")

# Comparing the two models
print("\n--- Model Comparison ---")
print("Linear Regression vs Support Vector Regression:")
print(f"Linear Regression MSE: {mean_squared_error(y_test, lr_pred):.4f}")
print(f"SVR MSE: {mean_squared_error(y_test, svr_pred):.4f}")
print(f"Linear Regression R¬≤: {r2_score(y_test, lr_pred):.4f}")
print(f"SVR R¬≤: {r2_score(y_test, svr_pred):.4f}")
print(f"Linear Regression Accuracy: {accuracy_score(y_test, lr_pred_class):.4f}")
print(f"SVR Accuracy: {accuracy_score(y_test, svr_pred_class):.4f}")

# Visual comparison of model performance
plt.figure(figsize=(10, 6))
metrics = ['MSE', 'R¬≤', 'Accuracy']
lr_scores = [mean_squared_error(y_test, lr_pred), r2_score(y_test, lr_pred), accuracy_score(y_test, lr_pred_class)]
svr_scores = [mean_squared_error(y_test, svr_pred), r2_score(y_test, svr_pred), accuracy_score(y_test, svr_pred_class)]

x = np.arange(len(metrics))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, lr_scores, width, label='Linear Regression')
rects2 = ax.bar(x + width/2, svr_scores, width, label='SVR')

ax.set_ylabel('Scores')
ax.set_title('Model Performance Comparison')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()

# Add values on top of bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.4f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.show()

# Alternative approach: Convert to proper classification task
print("\n--- Alternative Classification Approach ---")

# Logistic Regression as a proper classification approach
log_reg = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(max_iter=1000))
])
log_reg.fit(X_train, y_train)
log_reg_pred = log_reg.predict(X_test)

# Support Vector Classification
svc = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', SVC(kernel='rbf', probability=True))
])
svc.fit(X_train, y_train)
svc_pred = svc.predict(X_test)

print("\nLogistic Regression (Classification) Performance:")
print(f"Accuracy: {accuracy_score(y_test, log_reg_pred):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, log_reg_pred, target_names=label_encoder.classes_))

print("\nSupport Vector Classification Performance:")
print(f"Accuracy: {accuracy_score(y_test, svc_pred):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, svc_pred, target_names=label_encoder.classes_))

# Confusion Matrix for Logistic Regression
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, log_reg_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix - Logistic Regression')
plt.tight_layout()
plt.show()

# Confusion Matrix for SVC
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, svc_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix - Support Vector Classification')
plt.tight_layout()
plt.show()

# Cross-validation for more robust evaluation
print("\n--- Cross-Validation Results ---")
cv_lr = cross_val_score(log_reg, X, y, cv=5, scoring='accuracy')
cv_svc = cross_val_score(svc, X, y, cv=5, scoring='accuracy')

print(f"Logistic Regression CV Accuracy: {cv_lr.mean():.4f} ¬± {cv_lr.std():.4f}")
print(f"SVC CV Accuracy: {cv_svc.mean():.4f} ¬± {cv_svc.std():.4f}")

# Visualize cross-validation results
plt.figure(figsize=(10, 6))
plt.bar(['Logistic Regression', 'SVC'], [cv_lr.mean(), cv_svc.mean()], yerr=[cv_lr.std(), cv_svc.std()], capsize=10)
plt.title('Cross-Validation Accuracy')
plt.ylabel('Accuracy')
plt.ylim(0, 1)

# Add values on top of bars
for i, v in enumerate([cv_lr.mean(), cv_svc.mean()]):
    plt.text(i, v + 0.02, f'{v:.4f}', ha='center')

plt.show()

# Print final recommendations
print("\n--- Final Analysis and Recommendations ---")
print("Based on the results, the recommended model for sleep disorder classification is:")
if cv_svc.mean() > cv_lr.mean():
    print("Support Vector Classification with the following parameters:")
    print(f"- Kernel: {svc.named_steps['classifier'].kernel}")
    print(f"- C: {svc.named_steps['classifier'].C}")
    print(f"- Gamma: {svc.named_steps['classifier'].gamma}")
else:
    print("Logistic Regression with the following parameters:")
    print(f"- Max iterations: {log_reg.named_steps['classifier'].max_iter}")
    print(f"- C: {log_reg.named_steps['classifier'].C}")

print("\nTop factors influencing sleep disorders (based on model coefficients):")
if hasattr(lr_model, 'coef_') and lr_model.coef_.shape[0] == len(feature_names):
    for feature, importance in lr_importance.head(5).values:
        print(f"- {feature}: {importance:.4f}")
````
